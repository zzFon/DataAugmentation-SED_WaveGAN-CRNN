```python
import tensorflow as tf
from IPython.display import display, Audio
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
# Load the graph
tf.reset_default_graph()
saver = tf.train.import_meta_graph(r'C:\Users\43476\Desktop\2\gan\wavegan-master\train1\infer\infer.meta')
graph = tf.get_default_graph()
sess = tf.InteractiveSession()
saver.restore(sess, r'C:\Users\43476\Desktop\2\gan\wavegan-master\train1\model.ckpt-5236')

# Create 50 random latent vectors z
_z = (np.random.rand(50, 100) * 2.) - 1
# _z = (np.ones((50, 100)) * 2.) - 1
# _z = [[0.5 for i in range(100)] for i in range(50)]
# _z = (_z * 2.) - 1

# Synthesize G(z)
z = graph.get_tensor_by_name('z:0')
G_z = graph.get_tensor_by_name('G_z:0')
_G_z = sess.run(G_z, {z: _z})

# Play audio in notebook
for i in range(50):
    y = _G_z[i,:,0]
    display(Audio(y, rate=16000))
    print(np.shape(y))
    plt.plot(y)
    plt.show()
```
